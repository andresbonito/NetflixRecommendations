#Bibliotecas

```python
import pandas as pd
```

#Base de Credits

```python
credits = pd.read_csv('https://raw.githubusercontent.com/andresbonito/NetflixRecommendations/main/credits.csv', delimiter=',')
credits
```

#Base de Titles

```python
titles = pd.read_csv('https://raw.githubusercontent.com/andresbonito/NetflixRecommendations/main/titles.csv', delimiter=',')
titles
```

#Tratamento

##Tratamento dos Credits

```python
credits.info()
```

```python
characters = credits['character'].unique()

n = 0
for i in enumerate(characters):
  n += 1
print(n)
```

```python
credits.columns
```

```python
new_credits = credits.drop(['person_id'], axis=1)
new_credits
```

##Tratamentos dos Titles

```python
resultado = credits['id'].isin(titles['id'])
resultado.unique()
```

```python
imdb_score = titles['imdb_score']

n = 0
for i in enumerate(imdb_score):
  n += 1
print(n)
```

```python
new_titles = titles.drop(['age_certification', 'runtime', 'imdb_id', 'tmdb_popularity', 'tmdb_score'], axis=1)
new_titles
```

##Nova tabela

```python
credit_title = pd.merge(new_credits, new_titles, on='id')
credit_title
```

```python
movie_or_show = credit_title['type']
movie_or_show.unique()
```

```python
s = 0
m = 0
for i in movie_or_show:
  if i == 'MOVIE':
    m += 1
  elif i == 'SHOW':
    s += 1
print(f'H {m} movies e {s} shows')
```

```python
credit_title = credit_title.fillna(value= 'None')
credit_title.head(2)
```

```python
CreditTitleTable = credit_title.rename(columns={'name': 'Nome', 'character': 'Personagem', 'role': 'Papel', 'title': 'Titulo', 'type': 'Tipo', 'description': 'Descrio', 'release_year': 'Ano', 'genres': 'Gnero', 'production_countries': 'Pases filmados', 'seasons': 'Temporadas', 'imdb_score': 'Pontuao', 'imdb_votes': 'Votao'})
CreditTitleTable
```

```python
credit_title.to_excel('CreditTitle.xlsx', index=False)
```
